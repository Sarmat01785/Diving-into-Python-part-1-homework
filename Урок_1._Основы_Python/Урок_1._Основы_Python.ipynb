### Задача 1:

Треугольник существует только тогда, когда сумма любых двух его сторон больше третьей.

- Дано a, b, c - стороны предполагаемого треугольника. 
- Требуется сравнить длину каждого отрезка-стороны с суммой двух других. Если хотя бы в одном случае отрезок окажется больше суммы двух других, то треугольника с такими сторонами не существует.

- Отдельно сообщить является ли треугольник разносторонним, равнобедренным или равносторонним.

Решение задачи 1 вариант 1:

```python
# Вариант 1

# Оборачиваем блок кода в try-except для обработки возможных ошибок при вводе
try:
    # Запрашиваем у пользователя ввести первую сторону треугольника и преобразуем введенное значение в целое число
    a = int(input("Введите сторону треугольника a: "))
    # Запрашиваем у пользователя ввести вторую сторону треугольника и преобразуем введенное значение в целое число
    b = int(input("Введите сторону треугольника b: "))
    # Запрашиваем у пользователя ввести третью сторону треугольника и преобразуем введенное значение в целое число
    c = int(input("Введите сторону треугольника c: "))

    # Проверяем, являются ли введенные стороны a, b, c сторонами треугольника
    # Треугольник существует только тогда, когда сумма любых двух его сторон больше третьей
    # и каждая сторона больше нуля
    if a + b <= c or a + c <= b or b + c <= a or a <= 0 or b <= 0 or c <= 0:
        # Выводим сообщение, если треугольник не может существовать
        print("Треугольника с такими сторонами не существует")

    # Если треугольник существует, проверяем его тип
    elif a == b == c:
        # Все стороны равны, треугольник равносторонний
        print("Треугольник равносторонний")
    elif a == b or b == c or a == c:
        # Хотя бы две стороны равны, треугольник равнобедренный
        print("Треугольник равнобедренный")
    else:
        # Стороны не равны, треугольник разносторонний
        print("Треугольник разносторонний")
# Если при вводе возникла ошибка (например, пользователь ввел не число), перехватываем исключение
except ValueError:
    # Выводим сообщение об ошибке
    print("Введены некорректные данные")
```

Решение задачи 1 вариант 2:

```python
# Вариант 2

# Оборачиваем блок кода в try-except для обработки возможных ошибок при вводе

try:
    # Эта строка кода использует функцию map для преобразования ввода пользователя в целые числа.
    # Функция input позволяет пользователю ввести данные, а map применяет функцию int к каждому элементу списка, полученного после разделения ввода пользователя по пробелам.
    a, b, c = map(int, input("Введите стороны предполагаемого треугольника через пробел a, b, c: ").split())

    # Здесь мы проверяем, удовлетворяют ли введенные значения условиям существования треугольника.
    if a + b <= c or a + c <= b or b + c <= a or a <= 0 or b <= 0 or c <= 0:
        print("Треугольника с такими сторонами не существует")

    # Проверяем, является ли треугольник равносторонним.
    elif a == b == c:
        print("Треугольник равносторонний")

    # Проверяем, является ли треугольник равнобедренным.
    elif a == b or b == c or a == c:
        print("Треугольник равнобедренный")

    # Если ни одно из предыдущих условий не выполнено, треугольник считается разносторонним.
    else:
        print("Треугольник разносторонний")
# Если при вводе возникла ошибка (например, пользователь ввел не число), перехватываем исключение
except ValueError:
    # Выводим сообщение об ошибке
    print("Введены некорректные данные")
```

Решение задачи 1 вариант 3:

```python
# Вариант 3

# Этот код предназначен для определения типа треугольника по введенным значениям его сторон.

try:
    # Эта строка кода использует список comprehension для преобразования ввода пользователя в целые числа.
    # Функция input позволяет пользователю ввести данные, а split разделяет ввод пользователя по пробелам.
    a, b, c = [int(x) for x in input("Введите стороны треугольника a, b, c через пробел: ").split()]

    # Здесь мы проверяем, все ли стороны треугольника являются положительными числами и не равны нулю.
    if a <= 0 or b <= 0 or c <= 0:
        print(
            "Все стороны треугольника должны быть положительными числами и не должны быть равны нулю."
        )

    # Проверяем, удовлетворяют ли введенные значения условиям существования треугольника.
    elif a + b <= c or a + c <= b or b + c <= a:
        print("Треугольника с такими сторонами не существует")

    # Проверяем, является ли треугольник равносторонним.
    elif a == b == c:
        print("Треугольник равносторонний")

    # Проверяем, является ли треугольник равнобедренным.
    elif a == b or b == c or a == c:
        print("Треугольник равнобедренный")

    # Если ни одно из предыдущих условий не выполнено, треугольник считается разносторонним.
    else:
        print("Треугольник разносторонний")

except ValueError:
    # В случае возникновения ошибки при преобразовании введенных данных в целые числа, выводится сообщение об ошибке.
    print("Введены некорректные данные. Убедитесь, что вводятся три целых числа.")
```

### Задача 2:

Напишите код, который запрашивает число и сообщает является ли оно простым или составным. 
Используйте правило для проверки: “Число является простым, если делится нацело только на единицу и на себя”. 
Сделайте ограничение на ввод отрицательных чисел и чисел больше 100 тысяч.

Решение задачи 2 вариант 1

```python
# Запрашиваем у пользователя число
num = int(input("Введите число от 1 до 100000: "))

# Проверяем, является ли число отрицательным или больше 100000
if num <= 0 or num > 100000:
    print("Число вне допустимого диапазона.")
# Проверяем, не является ли число 1 составным или простым, так как 1 не простое и не составное
elif num == 1:
    print("1 не является ни простым, ни составным числом.")
else:
    # Проверяем, есть ли делители у числа, кроме 1 и самого числа
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    
    # Выводим результат
    if is_prime:
        print(f"Число {num} является простым.")
    else:
        print(f"Число {num} является составным.")
```

Решение задачи 2 вариант 2

```python
# Определяем функцию для проверки, является ли число простым
def is_prime(number):
    """
    Это docstring функции, описывающий ее поведение, аргументы и возвращаемое значение.
    Функция is_prime проверяет, является ли число простым. Число считается простым, если
    оно больше 1 и не имеет делителей, кроме 1 и самого себя.
    """
    # Если число меньше или равно 1, оно не может быть простым или составным
    if number <= 1:
        return False
    # Перебираем возможные делители числа от 2 до квадратного корня из числа
    for i in range(2, int(number ** 0.5) + 1):
        # Если число делится нацело, то оно составное
        if number % i == 0:
            return False
    # Если ни один делитель не найден, число простое
    return True

# Основной блок кода, который пытается выполнить следующие инструкции
try:
    # Запрашиваем у пользователя ввести число и преобразуем введенную строку в целое число
    number = int(input("Введите число в диапазоне от 1 до 100000: "))
    # Проверяем, равно ли число 1
    if number == 1:
        # Выводим сообщение, что 1 не является ни простым, ни составным числом
        print("1 не является ни простым, ни составным числом.")
    # Проверяем, находится ли число вне допустимого диапазона
    elif number <= 0 or number > 100000:
        print("Число вне допустимого диапазона.")
    # Если число в допустимом диапазоне, проверяем, является ли оно простым
    elif is_prime(number):
        print(f"Число {number} является простым.")
    # Если число не является простым, оно составное
    else:
        print(f"Число {number} является составным.")
# Если при преобразовании ввода в целое число произошла ошибка, выводим сообщение об ошибке
except ValueError:
    print("Введены некорректные данные. Необходимо ввести целое число.")
```

### Задача 3:

Программа загадывает число от 0 до 1000. Необходимо угадать число за 10 попыток.  Программа должна подсказывать “больше” или “меньше” после каждой попытки. 
Для генерации случайного числа используйте код:
from random import randintnum = randint(LOWER_LIMIT, UPPER_LIMIT)


Решение задачи 3

```python
# Импортируем модуль random для генерации случайных чисел
import random

# Генерируем случайное число от 0 до 1000
num = random.randint(0, 1000)

# Определяем количество попыток
COUNT = 10
# Определяем нижнюю границу диапазона чисел
LOWER_LIMIT = 0
# Определяем верхнюю границу диапазона чисел
UPPER_LIMIT = 1000

# Пока количество попыток больше нуля, выполняем следующие шаги
while COUNT > 0:
    # Попросим пользователя ввести число
    try:
        # Преобразуем введенное значение в целое число
        user_num = int(input(f"Программа загадала число от {LOWER_LIMIT} до {UPPER_LIMIT}. У вас {COUNT} попыток. Введите ваше число: "))

        # Проверяем, совпало ли введенное число с загаданным
        if user_num == num:
            print("Вы угадали число!")
            break

        # Если введенное число больше загаданного, обновляем верхнюю границу
        if user_num > num:
            print("Загаданное число меньше.")
            UPPER_LIMIT = user_num
        # Если введенное число меньше загаданного, обновляем нижнюю границу
        else:
            print("Загаданное число больше.")
            LOWER_LIMIT = user_num

        # Уменьшаем количество попыток на 1
        COUNT -= 1

    # Если пользователь ввел не целое число, сообщаем ему об ошибке
    except ValueError:
        print("Пожалуйста, введите целое число.")

# Если пользователь исчерпал все попытки, сообщаем ему загаданное число
if COUNT == 0:
    print(f"Вы исчерпали все попытки. Загаданное число было: {num}.")
```
